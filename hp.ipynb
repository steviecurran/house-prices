**Pre-processing**
- List and remove the features with missing values. 'Electrical' has only one missing value, but is categorical and is unevenly distributed
- Check the number of unique values. If this is low, feature may not be worth keepping.
- Get dummies for the categorical features (including 'MSSubClass')

```python
def prep(data):
    #print(data['Id'])
    # MISSING VALUES
    cols = data.columns; #print(cols)
    elec = data['Electrical'].unique(); 
    
    for (i,E) in enumerate(elec):
        tmp = data[data['Electrical']== E]; 
        #print(len(tmp)) # ONLY 1 MISSING BUT NOT EVENLY DISTRIBUTED, SO DROP
    
    for (i,col) in enumerate(cols):
        if data[col].isnull().sum().any() > 0:    
            print("%s  - missing values = " %(col), data[col].isnull().sum())
            del data[col]
     
    del data['Id']
    cols = data.columns;
    '''      
    
    for (i,col) in enumerate(cols):
        print("%s  - unique values = %d" %(col,len(data[col].unique()))) # ALL > 1
    '''
    
    num = data.select_dtypes(include='number'); 
    # MSSubClass CATEGORICAL
    nums = num.columns
    #del data['Id']
    cats = list(set(cols) - set(nums))
    cats.append('MSSubClass'); #print(cats)
    data= pd.get_dummies(data,columns=cats); #print(data) #230 columns
 
    return data
df_train = prep(df)
df_train
```

Note that the above is for training and validating this data - gives 230 fields. Have alternate code for training data where the dummies result in a different number of variables (as in the [Kaggle compeitition](https://www.kaggle.com/competitions/home-data-for-ml-course/overview)).

```python
from sklearn.model_selection import train_test_split,RandomizedSearchCV,GridSearchCV
from sklearn.preprocessing import StandardScaler
import xgboost as xg
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error,r2_score 
from matplotlib import cm
from scipy.stats import gaussian_kde, binned_statistic_2d,randint
```

Training and validation of the model. Typically reaching a regression coefficient of *r* > 0.9

```python
def ML(data):
    target = 'SalePrice'
    test_frac = 0.2; 
    data = data.sample(frac=1) # SHUFFLE
    data[target] = np.log10(data[target])

    X = data.drop([target], axis = 1); y = data[target]; 
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_frac, random_state=42)

    X_to_scale=X_train.loc[:, ~X_train.columns.isin([target])]
    cols_to_scale = X_to_scale.columns; 
    scaler = StandardScaler()
    scaler.fit(data[cols_to_scale])
    X_train[cols_to_scale] = scaler.transform(X_train[cols_to_scale])
    X_test[cols_to_scale] = scaler.transform(X_test[cols_to_scale])

    regressors = {
        "RF":  RandomForestRegressor(max_depth = 17, n_estimators = 216),
        "XGB": xg.XGBRegressor(colsample_bytree = 0.7, learning_rate = 0.03,
                                max_depth = 5, min_child_weight = 4,
                                n_estimators = 500, nthread = 4,
                                objective = 'reg:squarederror',subsample = 0.7),
        #"kNN": KNeighborsRegressor(n_neighbors = 10, weights='uniform'),
        #"DTR": DecisionTreeRegressor(max_depth = 10),
    }
    for key, model in regressors.items():
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test) # making predictions
        rms = mean_squared_error(y_test, y_pred)# , squared=False)
        mae = mean_absolute_error(y_test, y_pred)
        r2 = r2_score(y_test, y_pred) # Coefficient of Determination

        print("For %s (%d out of %d houses), RMS =  %1.3e MAE = %1.3e r = %1.3f"
              %(key,len(y_test),len(data),rms, mae,r2**0.5))

ML(df_train)
```
